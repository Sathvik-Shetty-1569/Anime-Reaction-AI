```python
import cv2
imgpath = "reaction_images/angry.jpg"
img = cv2.imread(imgpath)
img = cv2.resize(img,(300,300))
cv2.imshow("Reaction Output",img)
cv2.waitKey(0)  # üëà keeps window open until key press
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print("‚ùå Webcam not detected!")
    exit()

while True:
    frame ,ret = cap.read()
    if not ret:
        print("‚ö† Can't receive frame. Exiting...")
        break

    cv2.imshow("Live Webcam", frame)

    # Press 'q' to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()

```
